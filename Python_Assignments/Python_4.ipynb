```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# loading the data

Link = "https://data.nasa.gov/resource/y77d-th95.json"
dataset = pd.read_json(Link, orient = "columns")
dataset.head(5)
```

```python
# dropping the unnecessary columns: 
dataset = dataset.drop(labels=[":@computed_region_nnqa_25f4", ":@computed_region_cbhk_fwbd"], axis=1)
dataset.head(5)
```

```python
dataset["year"] = dataset['year'].str[:4]
dataset.head(5)
```

```python
for i in range(len(dataset)):
    if pd.isnull(dataset["geolocation"][i]):
        continue
    else:
        dataset["geolocation"][i] = dataset["geolocation"][i]["coordinates"]

dataset.head(5)
```

```python
output_path = 'space.csv'
dataset.to_csv(output_path, index=False)

print("Jason Data loaded and saved as Excel successfully.")
```
